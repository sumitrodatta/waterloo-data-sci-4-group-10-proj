```python
import numpy as np
import pandas as pd
from datetime import datetime, time
```

```python
def time_of_day(i):
    if (i > time(hour=6)) &  (i < time(hour=11,minute=59,second=59)):
        return 'Morning'
    elif (i > time(hour=12)) &  (i < time(hour=17,minute=59,second=59)):
        return 'Afternoon'
    elif (i > time(hour=18)) &  (i < time(hour=23,minute=59,second=59)):
        return 'Evening'
    else:
        return 'Night'
```

```python
def compress_precip_data(file_name):
    data=pd.read_csv(file_name)
    data['date_converted']=[datetime.strptime(d, '%Y-%m-%dT%H:%M:%S') for d in data['date']]
    data['actual_date']=[d.date() for d in data['date_converted']]
    data['actual_time']=[d.time() for d in data['date_converted']]
    data['week']=[d.strftime("%W") for d in data['actual_date']]
    data['year']=[d.year for d in data['actual_date']]
    data['month']=[d.month for d in data['actual_date']]
    data['day']=[d.day for d in data['actual_date']]
    data['time_of_day']=[time_of_day(d) for d in data['actual_time']]
    data.drop(['date','date_converted','actual_date','actual_time'],axis=1,inplace=True)
    grouped=pd.DataFrame(data.groupby(['year','month','day','time_of_day','week'],as_index=False)['rainfall'].sum())
    return(grouped)
```

```python
result=compress_precip_data("precipitation-data-2019.csv")
```
