```python
import numpy as np
import pandas as pd
from datetime import datetime, time
```

```python
def time_of_day(i):
    if (i > time(hour=6)) &  (i < time(hour=11,minute=59,second=59)):
        return 'Morning'
    elif (i > time(hour=12)) &  (i < time(hour=17,minute=59,second=59)):
        return 'Afternoon'
    elif (i > time(hour=18)) &  (i < time(hour=23,minute=59,second=59)):
        return 'Evening'
    else:
        return 'Night'
```

```python
dataLinks= ["https://ckan0.cf.opendata.inter.prod-toronto.ca/dataset/f2933501-0373-4734-b50c-4e4f39646180/resource/232e6023-0688-4d5f-aaf3-c92edd08cf1b/download/precipitation-data-2019.csv", 
            "https://ckan0.cf.opendata.inter.prod-toronto.ca/dataset/f2933501-0373-4734-b50c-4e4f39646180/resource/2a444824-7cf3-4ff7-aa19-6e434c35c7e9/download/precipitation-data-2020.csv"]
```

```python
dfsPrecipitation=[[],[]]
for link in dataLinks:
    dfsPrecipitation[i] = pd.read_csv(link)
```

```python
def compress_precip_data(df):
    data=df.copy()
    data['date_converted']=[datetime.strptime(d, '%Y-%m-%dT%H:%M:%S') for d in data['date']]
    data['actual_date']=[d.date() for d in data['date_converted']]
    data['actual_time']=[d.time() for d in data['date_converted']]
    data['week']=[d.strftime("%W") for d in data['actual_date']]
    data['year']=[d.year for d in data['actual_date']]
    data['month']=[d.month for d in data['actual_date']]
    data['day']=[d.day for d in data['actual_date']]
    data['time_of_day']=[time_of_day(d) for d in data['actual_time']]
    data['hour']=[d.hour for d in data['actual_time']]
    data.drop(['date','date_converted','actual_date','actual_time'],axis=1,inplace=True)
    grouped=pd.DataFrame(data.groupby(['year','month','day','hour','time_of_day','week'],as_index=False)['rainfall'].sum())
    return(grouped)
```

```python
result=compress_precip_data(dfsPrecipitation[0])
```
